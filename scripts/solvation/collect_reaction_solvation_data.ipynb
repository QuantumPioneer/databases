```python
import pathlib

from functools import cache

import pandas as pd
import swifter

from rdkit import Chem
```

```python
NOTEBOOK_DIR = pathlib.Path().absolute()
DATA_DIR = NOTEBOOK_DIR.parents[1] / "data" / "solvation"
QUANTUM_GREEN_DIR = pathlib.Path("/home/shared/projects/quantum_green")
PAPER_DATA_DIR = QUANTUM_GREEN_DIR / "paper" / "data"
PAPER_FIGURE_DIR = QUANTUM_GREEN_DIR / "paper" / "figure" / "section_3_1_2_ts_char"
SUPERCLOUD_DIR = QUANTUM_GREEN_DIR / "supercloud_xfer" / "cosmo"
RESULTS_CSV_DIR = SUPERCLOUD_DIR / "ts_26may5" / "results" / "csv"
```

```python
@cache
def canonical_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    for atom in mol.GetAtoms():
        atom.SetAtomMapNum(0)
    return Chem.MolToSmiles(mol, isomericSmiles=False)


pd.set_option("display.max_columns", None)


def head(df, n=2):
    display(df.head(n))
    print(f"Contains {len(df)} rows")


def clean_rxn_smi(rxn_smi):
    return ">>".join(
        [
            ".".join([canonical_smiles(smi) for smi in category.split(".")])
            for category in rxn_smi.split(">>")
        ]
    )


def get_rxn_smi(row):
    return clean_rxn_smi(
        row["r1"] + "." + row["r2"] + ">>" + row["p1"] + "." + row["p2"]
    )
```

```python
solvents_df = pd.read_csv(pathlib.Path.cwd() / "solvents.csv")
solvents_df["smiles"] = solvents_df["smiles"].swifter.apply(canonical_smiles)
head(solvents_df)
```

## Transition State Data

```python
ts_data = pd.read_pickle(
    PAPER_FIGURE_DIR / "quantum_green_ts_data_24july2c.pkl"
)
ts_data["rxn_smi"] = ts_data["smiles"].swifter.apply(clean_rxn_smi)
head(ts_data)
```

```python
def as_str(x):
    return str(int(x)) if isinstance(x, float) else x


ts_mapping = dict(
    zip(ts_data["ts_dlpno_log_filename_id"].apply(as_str), ts_data["rxn_smi"])
)
```

```python
may5_data_all = pd.read_csv(RESULTS_CSV_DIR / "all_results.csv")
may5_data_all["solute_id"] = may5_data_all["solute_id"].astype(str)
head(may5_data_all)
```

```python
temperature = may5_data_all["T"].apply(int)
may5_data = may5_data_all[temperature == 298].reset_index(drop=True).copy()
entropy = (
    may5_data_all[temperature == 299]["Gsolv"].values
    - may5_data_all[temperature == 297]["Gsolv"].values
) / 2
may5_data["Hsolv"] = may5_data["Gsolv"] - 298.15 * entropy
may5_data["solvent_smiles"] = may5_data["solvent"].map(
    solvents_df.set_index("cosmo_name")["smiles"]
)
may5_data["smiles"] = may5_data["solute_id"].map(ts_mapping)
head(may5_data)
```

```python
del may5_data_all
```

## Species data

```python
def mixture(df, species, solvent):
    return df[species] + "." + df[solvent]
```

```python
species_data = pd.read_csv(
    PAPER_DATA_DIR / "solvation" / "FILTERED_DEDUPLICATED_full_data_v3.csv",
    low_memory=False,
)
species_data["solvent_smiles"] = species_data["solvent_smiles"].swifter.apply(
    canonical_smiles
)
species_data["smiles"] = species_data["solute_smiles"].swifter.apply(canonical_smiles)
species_data["mixture"] = mixture(species_data, "smiles", "solvent_smiles")
head(species_data)
```

```python
species_property_mapping = {
    prop: species_data.set_index("mixture")[f"{prop} (kcal/mol)"]
    for prop in ["Gsolv", "Hsolv"]
}
```

## Reaction data

```python
reactants_and_products_df = (
    may5_data["smiles"]
    .str.replace(">>", ".")
    .str.split(".", expand=True)
    .rename(columns={0: "r1", 1: "r2", 2: "p1", 3: "p2"})
)
head(reactants_and_products_df)
```

```python
def compute_delta(df, prop, r_or_p):
    return df[f"{prop}"] - df[f"{r_or_p}1_{prop}"] - df[f"{r_or_p}2_{prop}"]
```

```python
reaction_data = may5_data[
    ["smiles", "solute_id", "solvent_smiles", "Gsolv", "Hsolv"]
].merge(reactants_and_products_df, left_index=True, right_index=True)
for prop in ["Gsolv", "Hsolv"]:
    for component in ["r1", "r2", "p1", "p2"]:
        print(f"Mapping {component} {prop}")
        mix = mixture(reaction_data, component, "solvent_smiles")
        reaction_data[f"{component}_{prop}"] = mix.map(species_property_mapping[prop])
reaction_data["DDGsolv_forward"] = compute_delta(reaction_data, "Gsolv", "r")
reaction_data["DDGsolv_reverse"] = compute_delta(reaction_data, "Gsolv", "p")
reaction_data["DDHsolv_forward"] = compute_delta(reaction_data, "Hsolv", "r")
reaction_data["DDHsolv_reverse"] = compute_delta(reaction_data, "Hsolv", "p")
head(reaction_data)
```

```python
reaction_data.rename(
    columns={
        "smiles": "solute_smiles",
        "solute_id": "solute_name",
        "Gsolv": "Gsolv (kcal/mol)",
        "DDGsolv_forward": "DDGsolv_forward (kcal/mol)",
        "DDGsolv_reverse": "DDGsolv_reverse (kcal/mol)",
        "Hsolv": "Hsolv (kcal/mol)",
        "DDHsolv_forward": "DDHsolv_forward (kcal/mol)",
        "DDHsolv_reverse": "DDHsolv_reverse (kcal/mol)",
    }
)[
    [
        "solvent_smiles",
        "solute_name",
        "solute_smiles",
        "Gsolv (kcal/mol)",
        "Hsolv (kcal/mol)",
        "r1",
        "r2",
        "p1",
        "p2",
        "r1_Gsolv",
        "r2_Gsolv",
        "p1_Gsolv",
        "p2_Gsolv",
        "DDGsolv_forward (kcal/mol)",
        "DDGsolv_reverse (kcal/mol)",
        "r1_Hsolv",
        "r2_Hsolv",
        "p1_Hsolv",
        "p2_Hsolv",
        "DDHsolv_forward (kcal/mol)",
        "DDHsolv_reverse (kcal/mol)",
    ]
].to_pickle(NOTEBOOK_DIR / "may5_reaction_data.pkl")
```
